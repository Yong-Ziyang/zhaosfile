```python
# svm 处理数据
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
import sklearn
from sklearn import svm 
```

```python
dataset=pd.read_excel('C:/Users/admin/Desktop/DATA.xlsx')
dataexlabel=dataset.iloc[:,:36]
datalabel=dataset.iloc[:,36]
naexist=(dataset=='n.s.').sum(axis=1)
datasetexna=dataexlabel[naexist==0]
datalabelexna=datalabel[naexist==0]
```

```python
from sklearn.preprocessing import LabelBinarizer
labelbin=LabelBinarizer()
label=labelbin.fit_transform(datalabelexna)
#from sklearn.preprocessing import StandardScaler
#standsca=StandardScaler()
#datasetnom=standsca.fit_transform(datasetexna)
#datasetnom=DataFrame(datasetnom,columns=datasetexna.columns,index=datasetexna.index)
```

```python
# 划分测试集和训练集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(datasetexna, label, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'C':np.linspace(0.1,10,num=20),'gamma':np.linspace(0.00001,0.0004,num=20)}
svmgrid=GridSearchCV(svm.SVC(),parameters)
svmgrid.fit(X_train.iloc[:,:15],y_train)
```

```python
score=svmgrid.grid_scores_
```

```python
def getnum(score):
    from matplotlib import pyplot as plt
    import numpy as np
    from mpl_toolkits.mplot3d import Axes3D
    meanscore=[]
    c=np.linspace(0.1,10,num=20)
    gamma=np.linspace(0.00001,0.0004,num=20)
    c,gamma=np.meshgrid(c,gamma)
    for i in range(len(score)):
        meanscore.append(score[i].mean_validation_score)
    meanscore1=np.array(meanscore).reshape(20,20).T
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_surface(X=c,Y=gamma,Z=meanscore1)
    ax.set_title('model accuracy within the change of  c(0-10) and gamma(0.00001-0.0004)')
    ax.set_xlabel('c')  
    ax.set_ylabel('gamma')  
    ax.set_zlabel('accuracy')
    plt.show()
```

```python
getnum(score)
```
